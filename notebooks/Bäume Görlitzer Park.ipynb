```python
# necessary libraries
import pandas as pd
import os
import json
import urllib
import folium
```

```python
# takes a big amount of time, only run in case of data update
complete = False;
trees = [];

url = 'https://trees.codefor.de/api/trees/?address=G%C3%B6rlitzer%20Park'

while not complete:
    data_raw = json.load(urllib.request.urlopen(url))
    trees = trees + data_raw['features']
    if (data_raw['next']):
        url = data_raw['next']
    else:
        complete = True
with open('data/trees.json', 'w') as outfile:
    json.dump(trees, outfile)
    print('written file')
```

```python
with open('data/trees.json') as json_data:
    d = json.load(json_data)
    mapped = list(map(lambda x: x['properties'], d))
    df = pd.DataFrame(mapped)
    
```

```python
df['trunk_circumference'] = df['trunk_circumference'].astype('int')
df['tree_height'] = df['tree_height'].astype('int')
df
```
